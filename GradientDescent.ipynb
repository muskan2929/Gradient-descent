```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5])

y=np.array([2,4,6,8,10])
```

```python
plt.scatter(x,y)

plt.show()
```

```python
m=0
b=0
```

```python
lr=0.1
```

```python
epochs=1000
```

```python
losses = []

for i in range(epochs):
  y_pred=m*x+b
```

```python
error=y-y_pred
```

```python
loss=np.mean(error**2)
```

```python
losses.append(loss)
```

```python
dm=-(2/len(x))*np.sum(x*error)
```

```python
db=-(2/len(x))*np.sum(error)
m=m-lr*dm
b=b-lr*db
print("slope:",m)
print("bias:",b)
```

```python
plt.plot(losses ,color="red")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.title("Loss Reduction")

plt.show()
```

```python
y_pred=m*x+b

plt.scatter(x,y)

plt.plot(x,y_pred,color="red")

plt.show()
```
